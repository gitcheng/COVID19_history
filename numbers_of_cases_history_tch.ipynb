```python
import pandas as pd
from utilities import *
print(pd.__version__)
```

## 累積病例數量前十名

```python
data_confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
data_confirmed_c = merge_locals(data_confirmed)
```

```python
plot_cumulated_histories(data_confirmed_c, 0, 10, plot_remaining=True)
plt.title('Top 10 countries with the most confirmed cases', fontsize='xx-large');
```

## 累積病例數量十一到二十名

```python
plot_cumulated_histories(data_confirmed_c, 10, 20)
plt.title('Top 11-th to 20-th countries with the most confirmed cases', fontsize='xx-large');
```

## 死亡數前十名

```python
data_deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
data_deaths_c = merge_locals(data_deaths)
```

```python
plot_cumulated_histories(data_deaths_c, 0, 10, case='deaths', plot_remaining=True)
plt.title('Top 10 countries with the most deaths', fontsize='xx-large');
```

## 死亡數第十一名到二十名

```python
plot_cumulated_histories(data_deaths_c, 10, 20, case='deaths')
plt.title('Top 11-th to 20-th countries with the most deaths', fontsize='xx-large');
```
